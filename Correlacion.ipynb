### Relación - Altura / peso

Datos con relacion mas fuerte y en tendencia positiva

```python
plt.scatter(personas["altura"], personas["peso"])
plt.xlabel("Altura (CMS)")
plt.ylabel("Peso (KGS)")
plt.show()
```

### Relación: Ingreso y Horas Trabajadas

Datos mas dispersos, tendencia positiva

```python
plt.scatter(personas["ingreso"], personas["horas_trabajadas"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Horas Trabajadas")
plt.show()
```

### Relación: Ingreso y

Relación con relacion fuerte, tendencia negativa

```python
plt.scatter(personas["ingreso"], personas["ausencias"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Ausencias")
plt.show()
```

### Relación: Ingreso y Peso de una persona

No hay relacion, y por ende no hay tendencia

```python
plt.scatter(personas["ingreso"], personas["peso"])
plt.xlabel("Ingreso ($)")
plt.ylabel("Peso (kg)")
plt.show()
```

## Matriz de Correlación

- Tenemos un valor entre -1 y 1,
- Mas cercano al 1 o al -1 mas fuerte la relación
- Mas cera al 0 relacion debil
- Signo - marca la tendencia negativa, y el + tendencia positiva

```python
personas.corr()
```

### Mapa de calor de la correlación

```python
matriz = personas.corr()
plt.matshow(matriz)
plt.show()
```

```python
matriz = personas.corr()
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1) # cmap le indicamos el color (sus iniciales), de -1, 0 y 1 Cmin,Vmax coloca el valor minimo y maximo
plt.colorbar()
plt.xticks(range(5),personas.columns, rotation=90) # numero de rango y el nombre de las columnas segun nuestro dataset, rotation - rotacion de etiquetas
plt.yticks(range(5),personas.columns)

for i in range(len(matriz.columns)): # recorre filas
  for j in range(len(matriz.columns)): # Recorre columnas
    plt.text(i, j, round(matriz.iloc[i,j],2), ha="center", va="center")
    # plt.text coloca el texto de i y j, instruccion redondeamos, iloc ingresamos al valor. 2 = al redondeo que deseamos
    # ha - horizontal alignment, va - vertical alignment

plt.show()
```
