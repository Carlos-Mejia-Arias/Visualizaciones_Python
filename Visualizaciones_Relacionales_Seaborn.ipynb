# Seaborn
## relacionales
### scatterplot y linearplot

```python
import matplotlib.pyplot as plt
import seaborn as sns # es una especializacion de de matplotlib para mostrar graficas estadisticos

#sns.get_dataset_names() permite ver las librerias que trae cargado sns
pinguinos = sns.load_dataset('penguins')
pinguinos
pinguinos.columns = ["especie", "isla", "longitud_pico", "profundidad_pico", "longitud_aleta", "masa", "sexo"] # cambiamos los nombres a español
pinguinos
```

## Primera gráfica

```python
sns.set_theme(style="darkgrid", context="talk")
relacional = sns.relplot(data = pinguinos, height = 7, aspect = 1.2, markers = ["s", ".", "<", "*"]) # hacmos una variable para poder poner el set con mas personalizacion, aspect que tan achatada o alargada se ve la grafica
relacional.set(title="Análisis de pinguinos")
relacional.set_ylabels("Variables")
relacional.set_xlabels("Número de Pinguinos")
# plt.grid(False)  podemos cambiar el lienzo si usamos el plt
plt.show()
```

## Diagramas de dispersión

```python
print(">>>>>>>>>>>>>>>>>>>> SIMPLE")

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta")
```

```python
print(">>>>>>>>>>>>>>>>>>>> MATIZ Y TAMAÑO FIJO")

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="masa", # color asociado a alguna caracteristica
            s= 100) # tamaño del punto
```

```python
sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            s= 100)
```

```python
print(">>>>>>>>>>>>>>>>>>>> REJILLA, TAMAÑO PERSONALIZADO Y ESTILO")

sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            row="sexo", # row nos da segun la cantidad de valores en la variable una gfrafica para cada una
            size="masa", #tamaño de puntos segun la etiqueta
            sizes=(100,500), #tamaño personaliizado, el mas pequeño y luego el mas grande
            style="sexo")
```

```python
sns.relplot(data=pinguinos,
            x="longitud_pico",
            y="longitud_aleta",
            hue="especie",
            col="isla", # relacion de columnas en este caso por isla
            row="sexo",
            size="masa",
            sizes=(100,500),
            style="sexo")
```

## Graficos de lineas

```python
sns.relplot(data=pinguinos,
            x="especie",
            y="masa",
            kind="line",
            hue="sexo",
            style="sexo") #representa que la linea sea continua o discontinua
plt.show()
```

```python
sns.relplot(data=pinguinos,
            x="profundidad_pico",
            y="longitud_aleta",
            kind="line",
            hue="especie",
            style="especie")
plt.show()
```

Barras de error instrumento de la estadística
- intervalo de confianza (por defecto 95%)
si tomamos una muestra estaria siempre dentro de ese intervalo
podemos poner CI = 99 para añadir mas intervalo o poner ci= "sd" para mostrar la desviacion estandar ( que tan dispersos estan los datos)

- ya no se utiliza cid se usa errorbar

## Gráficas combinadas

### Lineas y dispersión

```python
figura, ejes = plt.subplots(1,2, figsize=(15,5)) # primer () = 1 lienzo con 2 figuras y tamaño en pulgadas

sns.lineplot(data=pinguinos,
           x="especie",
           y="masa",
           hue="sexo",
           style="sexo",
           errorbar="sd",
           ax= ejes[0]) # asignamos la posicion


sns.scatterplot(data=pinguinos,
           x="longitud_pico",
           y="longitud_aleta",
           hue="especie",
           s=300,
           ax=ejes[1])

plt.show()
```
