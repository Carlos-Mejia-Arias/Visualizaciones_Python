# Datos Categóricos a Visualizar

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="ticks", context="talk", palette="tab10")

consumos = sns.load_dataset("tips")

consumos.columns = ["total", "propina", "genero", "fumador",
                    "dia", "tipo", "comensales"]

consumos
```

# Gráficas categóricas de Dispersión

## Gráfica stripplot (por defecto)

```python
sns.catplot(data=consumos,
            x="dia",
            y="total",
            #jitter=False alinea los puntos
            hue="comensales", #puede ser numerica o numerica
            col= "fumador",
            s=10 # size da el tamaño de punto
            )
plt.show()
```

## Gráfica swarmplot



*   solo colocamos kind="swarm"
*   pone en la misma linea los valores que tienen mas o menos el mismo valor

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="swarm", hue="comensales", s=20, col="fumador")
plt.show()
```

## Gráficas categóricas de Distribución

### Gráfica boxplot

*   Kind = box

Representa:

*   Linea superior = valor Maximo
*   Caja superior = cuartil 3
*   Linea media = mediana
*   Caja inferior = cuartil 1
*   Linea inferior = valor minimo





```python
sns.catplot(data=consumos, x="dia", y="total",
            kind="box",
            col="fumador")
plt.show()
```

### Gráfica boxenplot

Solo cuando hay muchos datos

*   Kind = boxen


```python
sns.catplot(data=consumos, x="dia", y="total",
            kind="boxen",
            col="fumador")
plt.show()
```

### Gráfica violinplot



*   Kind = violin

Representacion:

*   interior = diagrama de caja
*   la forma es una forma aproximacion de los datos

```python
sns.catplot(data=consumos, x="dia", y="total",
            kind="violin",
            )
plt.show()
```

```python
sns.catplot(data=consumos, x="dia", y="total",
            kind="violin",
            hue="fumador"
            )
plt.show()
```

```python
sns.catplot(data=consumos, x="dia", y="total",
            kind="violin",
            hue="fumador",
            split=True #parte la mitad cada uno del hue y los pega
            )
plt.show()
```

## Gráficas categóricas de Estimación
Estiman el intervalo de confianza del promerio o la desviacion estandar

### Gráfica pointplot

*   kind="point"

Linea vertical que cruza los puntos es el intervalo de confianza al 95%

```python
sns.catplot(data=consumos, x="dia", y="total",
            kind="point",
            col="fumador",
            )
plt.show()
```

### Gráfica barplot

*   kind="bar"

```python
sns.catplot(data=consumos, x="dia", y="total",
            kind="bar",
            hue="comensales",
            )
plt.show()
```

### Gráfica countplot
Misma informacion que un histograma
*   kind="count"




```python
sns.catplot(data=consumos, x="dia",
            kind="count",
            hue="comensales",
            )
plt.show()
```
